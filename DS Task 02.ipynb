```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic_df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print(titanic_df.head())

# Display basic information about the dataset
print(titanic_df.info())

# Display summary statistics of the dataset
print(titanic_df.describe())

# Data Cleaning
# Check for missing values
print(titanic_df.isnull().sum())
```

# Data Cleaning

```python
# Fill missing values for 'age' with the median age
titanic_df['age'].fillna(titanic_df['age'].median(), inplace=True)

# Fill missing values for 'embarked' with the most common port
titanic_df['embarked'].fillna(titanic_df['embarked'].mode()[0], inplace=True)

# Drop 'deck' column due to a large number of missing values
titanic_df.drop(columns='deck', inplace=True)

# Drop rows with missing 'embark_town'
titanic_df.dropna(subset=['embark_town'], inplace=True)

# Confirm missing values are handled
print(titanic_df.isnull().sum())
```

# Exploratory Data Analysis (EDA)


```python
# Plot the distribution of 'age'
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of survivors and non-survivors
plt.figure(figsize=(10, 6))
sns.countplot(x='survived', data=titanic_df, palette='viridis')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of passengers by class
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=titanic_df, palette='viridis')
plt.title('Passenger Class Count')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of passengers by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data=titanic_df, palette='viridis')
plt.title('Gender Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between age and survival
plt.figure(figsize=(10, 6))
sns.boxplot(x='survived', y='age', data=titanic_df, palette='viridis')
plt.title('Age vs. Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Explore the relationship between class and survival
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', hue='survived', data=titanic_df, palette='viridis')
plt.title('Class vs. Survival')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between gender and survival
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='survived', data=titanic_df, palette='viridis')
plt.title('Gender vs. Survival')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```
